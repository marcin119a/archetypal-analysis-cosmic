# SIMILARITY IN COSMIC

```python
# COSMIC SIGNATURE

COSMIC_sig=pd.read_csv('../utils/COSMIC_SBS_GRCh37.txt',sep='\t')
```

```python
# artifacts

artifact_list=['SBS27','SBS43']
artifact_list.extend(['SBS'+str(arch) for arch in range(45,61)])
```

```python
# reordering mutational classes and remove archetypes

COSMIC_sig['Type']=COSMIC_sig['Type'].apply(lambda x : x[0]+x[2]+x[6]+'-'+x[4])
COSMIC_sig=COSMIC_sig.drop(columns=artifact_list)
```

```python
# similarity

sim=pd.DataFrame(cosine_similarity(COSMIC_sig.iloc[:,1:].T))
sim.columns=COSMIC_sig.columns[1:]
sim.index=COSMIC_sig.columns[1:]
```

```python
# clustermap
sns.set_style('dark')

g=sns.clustermap(sim,figsize=(14,20),xticklabels=True, yticklabels=True,cmap='coolwarm')
g.ax_row_dendrogram.set_visible(False)
g.ax_col_dendrogram.set_visible(False)

plt.setp(g.ax_heatmap.get_xticklabels(), rotation=90)
# Adjust to have the column labels not cut out
plt.subplots_adjust(bottom=0.3)
g.ax_row_dendrogram.set_visible(False)
# Use the dendrogram box to reposition the colour bar
dendro_box = g.ax_row_dendrogram.get_position()
dendro_box.x0 = (dendro_box.x0 + 2 * dendro_box.x1) / 3
g.cax.set_position(dendro_box)
# Move the ticks to the left 
g.cax.yaxis.set_ticks_position("left")
```

```python
# signature list = > 0.8

dfs=[]

for pair in list(itertools.combinations(COSMIC_sig.columns[1:], 2)):
    s_1=COSMIC_sig[pair[0]].ravel().reshape(1,-1)
    s_2=COSMIC_sig[pair[1]].ravel().reshape(1,-1)
    cos=cosine_similarity(s_1,s_2)[0][0]
    df=pd.DataFrame([pair[0],pair[1],cos]).T
    df.columns=['sig1','sig2','similarity']
    dfs.append(df)

df_simi_all=pd.concat(dfs).sort_values(by='similarity',ascending=False).reset_index(drop=True)
df_simi_all[df_simi_all.similarity>0.8]
```

```python
df_similarity=[]
stats=[] 
for (scenario,n_scenario) in zip(scenarios,range(5)):
    temp_sim=sim.loc[scenario,scenario]
    temp_sim=temp_sim.mask(np.eye(temp_sim.shape[0], dtype = bool))
    df=pd.DataFrame(temp_sim.values.ravel())
    df['Scenario']=n_scenario+1
    df.columns=['Similarity','Scenario']
    df_similarity.append(df['Similarity'])

df_similarity.append(pd.DataFrame(np.array(sim.mask(np.eye(sim.shape[0], dtype = bool))).ravel())) # full set 
```

```python
list_scenario = [[i+1]*len(df_similarity[i]) for i in range(0,6)]
list_scenario = [item for sublist in list_scenario for item in sublist]
df_p=pd.concat(df_similarity)
df_p['scenario']=list_scenario
df_p.columns=['Cosine similarity\n','\nScenario']
df_p.loc[df_p['\nScenario']==6,'\nScenario']='Full set'
```

FLATNESS DEFINITION

```python
def flatness_measure(sig):  
    s_1=np.array([1/96]*96).reshape(1,-1)
    s_2=sig.ravel().reshape(1,-1)    
    return cosine_similarity(s_1,s_2)[0][0]

df_flats=[]
for scenario in range(0,5):
    flatness=[flatness_measure(COSMIC_sig[scenarios[scenario]][sig]) for sig in scenarios[scenario]]
    df=pd.DataFrame(flatness)
    df_flats.append(df)
    #print(f'flattness scenario {scenario+1}: ', np.median(flatness))

list_scenario = [[i+1]*len(df_flats[i]) for i in range(5)]
list_scenario = [item for sublist in list_scenario for item in sublist]
df_p=pd.concat(df_flats)
df_p['scenario']=list_scenario
df_p.columns=['Flatness\n','\nScenario']

plt.figure(figsize=(10,10))
sns.violinplot(data=df_p,y='Flatness\n',x='\nScenario');
```

```python
#Flatness distribution in COSMIC

fl=[[sig,flatness_measure(COSMIC_sig[sig])] for sig in COSMIC_sig.columns[1:]]
df_flatness=pd.DataFrame(fl,columns=['Signatures','Flatness']).sort_values(by='Flatness',ascending=False).reset_index(drop=True)

fig,ax=plt.subplots(1,1,figsize=(8,8))
sns.histplot(df_flatness['Flatness'],bins=15,ax=ax);
plt.xlabel('\nFlatness')
plt.ylabel('Count')
plt.ylim(0,12);
```

# ARCHETYPAL ANALYSIS

Extracting archetypes from 1 to 40 and select n_archetypes with 95% of the explained variance

```python
var_rss=[]

print('Extracting Archetypes ...')
for n_arch in range(1,40):
    AA = ArchetypalAnalysis(n_archetypes = n_arch,initialize = 'random',
                            redundancy_try = 30,tolerance = 0.0001, 
                            max_iter = 500, random_state = 45, C = 1)
    AA.fit(np.array(COSMIC_sig.iloc[:,1:]).T)
    var_rss.append(AA.explained_variance_)
```

Plot the explained variance vs number of archetypes

```python
sns.set(font_scale=.9)
fig, (ax1) = plt.subplots(1, 1, figsize=(10,8),sharex=True)

ax1.plot(range(1, 40), var_rss,'bo-',color='tab:blue')
ax1.set_xlabel('\n Number of Archetypes',fontsize=12)
ax1.set_ylabel('Explained variance \n',fontsize=12) 
plt.xticks(np.arange(1,40))
plt.yticks(np.arange(0,1.05,0.05))
ax1.set_ylim(0.,1.0)
ax1.set_xlim(0.5,40);ax1.axvline(29,0,1,color='orange',linestyle='-',lw=1.)
ax1.axhline(0.95,0,1,color='orange',linestyle='-',lw=1.);
```

Relationship among archetypes and signatures

```python
arch=pd.DataFrame(AA.archetypes)
arch.columns=['ARCH_'+str(num) for num in range(1,AA.archetypes.shape[1]+1)]
dictionaire={}
for column in COSMIC_sig.iloc[:,1:].columns:
    for column_2 in arch.columns:
        s_1=COSMIC_sig[column].ravel().reshape(1,-1)
        s_2=arch[column_2].ravel().reshape(1,-1)
        cos=cosine_similarity(s_1,s_2)
        if column!=column_2 and cos>0.8:
            dictionaire[str(cos[0][0])]=[column,column_2]
            
dfs=[]
for key in set(dictionaire.keys()):
    df=pd.DataFrame(dictionaire[key]).T
    df['similarity']=float(key)
    df.columns=['sig1','sig2','similarity']
    dfs.append(df)
    
df=pd.concat(dfs).sort_values(by='similarity',ascending=False).reset_index(drop=True)
```

```python
df.columns=['Signatures','Archetypes','Similarity']
```

```python
r_=[]
f_rec=[]
thresholds=np.linspace(0.8,1,20)

for threshold in thresholds:
    
    pure_sig=df[df['Similarity']>threshold]
    
    p=(pure_sig.iloc[:,:-1].Archetypes.value_counts()>=1)
    r=len(p[p==True])
    r_.append(r)
        
    f_rec.append(len(rec_profiles[rec_profiles['Cosine Similarity']>threshold]))


sns.set_style('dark')
sns.set(font_scale=1.1)
fig, (ax1) = plt.subplots(1, 1, figsize=(10,8),sharex=True)

ax1.plot(thresholds, r_,'bo-',color='tab:orange')

ax1.yaxis.set_ticks(np.arange(0,32,2))
ax1.set_ylabel('N of archetypes associated to at least one signature  \n',fontsize=12) 
ax1.set_xlabel('\n Cosine Similarity',fontsize=12);
```

```python
sim_profiles=[]

for s1,s2 in zip(df_simi_all.loc[:,'sig1'],df_simi_all.loc[:,'sig2']):
    s1_prof=contrib.loc[s1,:].values.ravel().reshape(1,-1)
    s2_prof=contrib.loc[s2,:].values.ravel().reshape(1,-1)
    sim_profiles.append(cosine_similarity(s1_prof,s2_prof)[0][0])
    
df_simi_all['cosine similarity alpha coefficients \n']=sim_profiles
```

```python
category = [(df_simi_all.similarity<0.2),
             ((df_simi_all.similarity>=0.2)&(df_simi_all.similarity<0.4)),
             ((df_simi_all.similarity>=0.4)&(df_simi_all.similarity<0.6)),
             ((df_simi_all.similarity>=0.6)&(df_simi_all.similarity<0.8)),
             ((df_simi_all.similarity>=0.8)&(df_simi_all.similarity<=1))]

values = ['$cos<0.2$','$0.2\leq\cos < 0.4$','$0.4\leq\cos < 0.6$','$0.6\leq\cos < 0.8$','$cos\geq\ 0.8$']

df_simi_all['\n cosine similarity Signatures']=np.select(category,values)
```
